# Veremos como usar el modelo entrenado en el script "c0_K_means.ipynb" para hacer predicciones sin volver a entrenar el modelo

```python
# Importamos funci贸n para usar el modelo entrenado
import joblib

# Importamos funci贸n para mdeir el accuracy:
from sklearn import metrics

# Importamos el dataset "make_blobs" que genera un conjunto de "manchas" 
from sklearn.datasets import make_blobs

import matplotlib.pyplot as plt 
```

```python
# Cargamos el modelo entrenado:
clf = joblib.load('modelo_entrenado_K_means_0.pkl')
```

```python
# Generamos los datos con el generador de datos 'make blobs':

K_features=10#<--- numero de centroides 'k' del K-means para generar los datos

n=1000 # <--- numero de datos

features, labels = make_blobs(n_samples=n, centers=K_features, random_state=42)

# Obs: random_state=42 genera datos deterministicos (ie. no aleatorios)
```

```python
# Generamos predicciones con el modelo entrenado:
labels_predict = clf.predict(features)
```

```python
# Calculamos la precisi贸n de la clasificaci贸n:
accuracy = metrics.adjusted_rand_score(labels , labels_predict)
accuracy
```

```python
# Graficamos un diagrama de dispersi贸n de los datos:
plt.figure()
plt.scatter(features[:,0], features[:,1], s=50) #<--- 's' indica el grosor de los puntos
plt.title('Features')
plt.show()
```

```python
# graficamos datos y un color 'c' correspondiente a su etiqueta
plt.scatter(features[:,0], features[:,1], c=labels_predict)

plt.title('Clasificaci贸n de Features')
plt.show()
```
